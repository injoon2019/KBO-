```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\injoo\광주 인공지능 사관학교\웜업 프로젝트\KBO 데이터 시각화\Dataset\game_schedule.csv')
```

```python
df = df.drop(columns = ['Unnamed: 0'])
```

```python
df
```

```python
df['Date'] = df["Date"].apply(lambda x: "20"+x[:2]+x[3:5])
```

```python
df.drop(columns=['Month', 'PlaceName'])
```

```python
def change_home(home):
    if home=='LG':
        return 'LGT'
    
    elif home=='SK':
        return'SKW'
    
    elif home =='KT':
        return 'KTW'
    
    elif home =='KIA':
        return 'KIA'
    
    elif home =='두산':
        return 'DSB'
    
    elif home =='키움':
        return 'KWH'
    
    elif home =='롯데':
        return 'LTG'
    
    elif home =='한화':
        return 'HHE'
    
    elif home =='NC':
        return 'NCD'
    
    elif home =='삼성':
        return 'SSL'
```

```python
df['HomeName'] = df['HomeName'].apply(lambda x: change_home(x))
```

```python
df['AwayName'] = df['AwayName'].apply(lambda x: change_home(x))
```

```python
df.to_csv(r'./game_schedule_for_strikes.csv')
```
